```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t=np.linspace(0,2*np.pi,100)#creating array for value of x
y=np.sin(4*np.pi*t)#defining function
import scipy.fftpack as fft
f=fft.fft(y)#amplitude calculation for each datapoint in frequency space
freq= fft.fftfreq(100,t[1]-t[0])#frequency calculation
plt.plot(freq,f)
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
```

2.)Test case for pyrotd

```python
import pandas as pd
from pyrotd import calc_spec_accels
```

```python

osc_damping=0.05#default value for damping
osc_freqs=np.logspace(-1,1,1000)#creating array for natural frequency
accl=np.ones_like(osc_freqs)#creating an array for acceleration with all values equal to 1
```

```python
spec_accels=calc_spec_accels(0.05,accl,osc_freqs,osc_damping)
```

```python
dataframe=pd.DataFrame(spec_accels)#creating a dataframe for ease of access
N_Freq=dataframe.iloc[::,0]#storing values of Natural frequency in a list
Peak_acc=dataframe.iloc[::,1]#storing values of  peak acceleration in a list
```

```python
plt.plot(N_Freq,Peak_acc)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Natural Frequency(HZ)')
plt.ylabel('Peak Acceleration(G)')
```

Interpretation of graph: When constant acceleration is applied the peak  acceleration will be equal to 1 
for small frequencies .

```python
osc_damping=0.05#default value for damping
osc_freqs=np.logspace(-1,1,1000)#creating array for natural frequency
#creating a sine pulse as input
t=np.linspace(0,10,1000)
omega=0.01
Theta=np.cumsum(omega*t)#theta =integral of angular velpcity wrt time,in discrete case this will be equal to cumulaitve 
sine_Pulse=np.sin(Theta)
list1=list(sine_Pulse)
for i in range(750):
    list1.append(0);
#Creating array for final signal
signal_Input=np.array(list1);
```

```python
spec_accels=calc_spec_accels(0.05,signal_Input,osc_freqs,osc_damping)
dataframe=pd.DataFrame(spec_accels)#creating a dataframe for ease of access
N_Freq=dataframe.iloc[::,0]#storing values of Natural frequency in a list
Peak_acc=dataframe.iloc[::,1]#storing values of  peak acceleration in a list
```

```python
plt.plot(N_Freq,Peak_acc)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Natural Frequency(HZ)')
plt.ylabel('Peak Acceleration(G)')
```
